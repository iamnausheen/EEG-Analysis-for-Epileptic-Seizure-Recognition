```python
# Correlation analysis

variables = eeg_df.loc[:, eeg_df.columns != "Data_item ID"]
correlation = variables.corr()["Category"]
correlation
```

```python
# None of the individual data points are correlated with the final category, it is how the data points continue throughout 
# time that represents brain activity.
```

```python
var_correlation = variables.corr()
var_correlation
```

```python
# The value of each variable is highly correlated to the ones close to it or around it and it decreases continuously the farther the variables get
# For ex: the correlation of X16 is 095 with X15, 0.823 with X14, 0.646 with X13, 0.45 with X12 and continues to decrease so on.
# correlation of X16 with X17 - 0.94, X18 - 0.77, X19 - 0.55, X20 - 0.35 and continues to decrease so on.
# Brain activity at a particular time is dependent on the activity going on just before, it follows a rhythmic continuous pattern
# Anomalies in such pattern, arythmic activity is what is linked to seizures
```

```python
# Visualisation libraries

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualising correlation among all the variables
subset = var_correlation.iloc[:30, :30]
plt.figure(figsize = (25, 12))
sns.heatmap(subset, annot=True, cmap="coolwarm")
plt.title("Correlation of Variables with Each Other", fontsize = 20)
plt.show()
```

```python
# The values around the diagnol values are highly correlated, correlation decreases on moving away from the diagnol values
```

```python
# Subplots for each category of activity, any random row has been selected with the corresponding category

colors = ["magenta", "gold", "green", "purple", "blue"]
rows_left = [1, 7, 12, 18, 48]
rows_right = [23, 45, 35, 46, 2]
fig, axes = plt.subplots(nrows = 5, ncols = 2, figsize = (10, 15))

# One second is divided in 178 points
x = []
time = 0
for j in range(0, 178):
    x.append(time)
    time = time + 1/178

for i, row_index in enumerate(rows_left):
    ax = axes[i, 0]
    y = eeg_df.iloc[row_index, 1:179]
    ax.plot(x, y, color=colors[i])
    ax.set_title(f"Category {i+1}")
    ax.set_xlabel("Time")
    ax.set_ylabel("Signal")

for i, row_index in enumerate(rows_right):
    ax = axes[i, 1]
    y = eeg_df.iloc[row_index, 1:179]
    ax.plot(x, y, color=colors[i])
    ax.set_title(f"Category {i+1}")
    ax.set_xlabel("Time")
    ax.set_ylabel("Signal")

plt.tight_layout()
plt.show()
```

```python
# Category 1 has much flatter, blunt waves - arythmic activity - indicative of seizures
# Category 3 at the left has an eye blink artifact at 0.2s
```

```python
# Visualising each category together for comparison
colors = ["magenta", "gold", "green", "purple", "blue"]
rows_left = [1, 7, 12, 18, 48]

for i in rows_left:
    y = eeg_df.iloc[i, 1:179]
    plt.plot(x, y, color = colors[rows_left.index(i)], label=f"Category {rows_left.index(i) + 1}")
    
plt.xlabel("Time")
plt.ylabel("Signal")
plt.legend()
plt.title("Comparing wave of each category")
plt.show()
```

```python
# Major difference can be observed between the seizure and non-seizure category, however it is not as evident on the EEG 
# itself making it difficult to identify
```

```python
from scipy.stats import norm

fig, axes = plt.subplots(nrows = 5, ncols = 3, figsize=(10, 10))

N_COLS = 3
k = 1
for j in range(0, N_COLS):
    feature_names = eeg_df.columns[k:k+5]
    for i, feature_names in enumerate(feature_names):
        ax = axes[i, j]
        data_f = eeg_df[feature_names]
        sns.histplot(data_f, kde=False, stat="density", ax = axes[i, j], bins=50, color="purple", label="Histogram")
        mu, std = norm.fit(data_f)
        xmin, xmax = ax.get_xlim()
        x = np.linspace(xmin, xmax, 100)
        p = norm.pdf(x, mu, std)
        ax.plot(x, p, 'y', linewidth=2, label="Gaussian Fit")
        ax.set_title(f"X{k}")
        ax.set_xlabel("Feature Value")
        ax.set_ylabel("Density")
        k = k+1

plt.tight_layout()
plt.show()
```
